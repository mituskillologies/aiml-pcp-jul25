```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##### Line plot

```python
x = np.array([7,12,15,22,27,45,56,61,72,78])
y = np.array([10,20,22,34,47,55,51,68,81,85])
z = np.array([78,77,76,56,54,37,43,23,21,12])
p = np.array([67,45,53,47,65,26,46,58,74,35])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title('My first plot')
plt.grid()
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.plot(x, y, color = 'green', marker = 'o')
```

```python
plt.title('My first plot')
plt.grid()
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.plot(x, y, color = 'green', marker = 'o', label = 'First')
plt.plot(x, z, color = 'brown', marker = 'o', label = 'Second')
plt.legend()
```

##### Scatter plot

```python
plt.title('My first plot')
plt.grid()
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.scatter(x, y, color = 'magenta', marker = '*', s = 100)
```

##### Bar plot

```python
df = pd.read_csv('student.csv')
```

```python
plt.title('Student Marks')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.bar(df['name'], df['marks'], color = 'navy')
```

```python
plt.title('Student Marks')
plt.ylabel('Names')
plt.xlabel('Marks')
plt.barh(df['name'], df['marks'], color = 'navy')
```

```python
plt.title('Student Marks')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.bar(df['name'], df['marks'], color = 'navy', width=0.5, 
        align='edge', label = 'Marks')
plt.bar(df['name'], df['age'], color = 'skyblue', width=0.5, 
        align='center', label = 'Age')
plt.legend()
```

##### Histogram

```python
marks = [67,65,45,86,74,56,61,66,73,96,58,62,73,75,81,48,64,56,58]
```

```python
plt.title('Histogram of Marks')
plt.xlabel('Range of Marks')
plt.ylabel('Counts')
plt.xticks(range(0,101,10))
plt.grid()
plt.hist(marks, bins = range(0,101,10), color = 'black');
```
