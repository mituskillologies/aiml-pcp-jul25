```python
import matplotlib.pyplot as plt
import numpy as np
```

##### Define data

```python
x = np.array([7,11,23,31,37,45,51,68,77,80])
y = np.array([12,18,22,35,38,44,59,66,71,75])
z = np.array([67,65,56,54,58,44,41,32,21,13])
p = np.array([78,56,45,65,73,27,65,47,56,53])
```

##### Line Plot

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color = 'red', marker = 'o')
```

```python
plt.title('Comparison plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color = 'red', marker = 'o', label = 'First')
plt.plot(x, z, color = 'green', marker = 'o', label = 'Second')
plt.legend()
```

##### Scatter Plot

```python
plt.title('Scatter plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.scatter(x, y, color = 'red', marker = 'o')
```

##### Histogram

```python
marks = [67,78,43,64,73,80,95,58,67,51,75,66,82,34,72,62,68]
```

```python
plt.title('Marks Histogram')
plt.xlabel('Range of Marks')
plt.ylabel('Count')
plt.grid()
plt.xticks(range(0,101,10))
plt.hist(marks, bins = range(0,101,10), color = 'magenta');
```

##### Boxplot

```python
plt.boxplot(marks);
```
