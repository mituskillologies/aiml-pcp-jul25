```python
import matplotlib.pyplot as plt
import numpy as np
```

##### Line plot

```python
x = np.array([7, 12, 15, 23, 37, 41, 55, 67, 71, 78])
y = np.array([8, 16, 22, 23, 27, 38, 45, 50, 73, 81])
z = np.array([67, 61, 56, 54, 42, 34, 22, 21, 26, 18])
p = np.array([56, 23, 47, 65, 37, 65, 69, 78, 72, 54])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color = 'red', marker = 'o')
```

##### Scatter Plot

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.scatter(x, y, color = 'red', marker = 'o')
```

##### Histogram

```python
marks = [67,73,45,57,69,73,55,89,82,57,66,68,53,70,72,34,93,64,65,72]
```

```python
plt.title("Distribution of Marks")
plt.xlabel('Marks Range')
plt.ylabel('Counts')
plt.xticks(range(0,101,10))
plt.grid()
plt.hist(marks, bins=range(0,101,10));
```
